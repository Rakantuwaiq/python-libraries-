# matplotlib

### import matplotlib 

```python
import matplotlib.pyplot as plt 
import numpy as np
```

## create a dummy data and plot it 

```python
y = [0,1,4,2,9]
x = [1,2,3,4,5]
```

```python
plt.plot(x,y)
```

### create a linear relationship between x and y 

```python
x = np.linspace(0,10, 20)
y = 2*x + 10
```

```python
x
```

```python
y
```

```python
plt.plot(x, y)
```

### create a quadratic relationship between x and y 

```python
x = np.linspace(0,10, 20)
y = x ** 2
plt.plot(x,y)
```

### create a exponential relationship between x and y 

```python
x = np.linspace(0,10, 20)
y = np.exp(x)
plt.plot(x,y)
```

### Colors

```python
plt.plot(x,y, 'g')
```

## labels

```python
plt.plot(x,y, 'c')
plt.xlabel('x data')
plt.ylabel('y data')
```

## markers

```python
plt.plot(x,y, 'c', marker ='o')
plt.xlabel('x data')
plt.ylabel('y data')
```

```python
plt.plot(x,y, 'c', marker= '*', ms = 4)
plt.xlabel('x data')
plt.ylabel('y data')
plt.show()
```

## title 

```python
plt.plot(x,y, 'c', marker ='o')
plt.xlabel('x data')
plt.ylabel('y data')
plt.title('my first plot')
```

```python
plt.plot(x,y, 'c', marker ='o')
plt.xlabel('x data')
plt.ylabel('y data')
plt.title('my first plot', loc='left')
```

## Grid

```python
plt.title('my first plot ')
plt.plot(x,y, 'c', label = 'one line')
plt.xlabel('x data')
plt.ylabel('y data')
plt.grid()
plt.legend()
```

## vertical grid

```python
plt.title('my first plot ')
plt.plot(x,y, 'c', label = 'one line')
plt.xlabel('x data')
plt.ylabel('y data')
plt.grid(axis= 'x')
plt.legend()
```

## horizontal grid 

```python
plt.title('my first plot ')
plt.plot(x,y, 'c', label = 'one line')
plt.xlabel('x data')
plt.ylabel('y data')
plt.grid(axis= 'y')
plt.legend()
```

## color and line width for the grid 

```python
plt.title('my first plot ')
plt.plot(x,y, 'c', label = 'one line')
plt.xlabel('x data')
plt.ylabel('y data')
plt.grid(color = 'blue', linewidth = 0.4)
plt.legend()
```

## Bar plot 

```python
categories = ['A','B','C','D'] 
values = [10, 8, 16, 20]

plt.bar(categories, values, color = 'green')
plt.title('bar plot')
plt.xlabel('categories')
plt.ylabel('values')
```

## Subplot

```python
x = np.linspace(0,10,100)
y1 = np.sin(x)
y2 = np.cos(x)

fig, axs = plt.subplots(2,1)

axs[0].plot(x,y1,'r')
axs[0].set_title('sin function')

axs[1].plot(x,y2, 'b')
axs[1].set_title('cos function')

plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```
